# Misura di esempio Rubidio

**Informazioni generali.** Misura della fotoluminescenza di un gas di atomi di Rubidio, eccitati tramite un laser accordabile in frequenza. Lo scopo dell'esperienza è di evidenziare la struttura fine e iperfine dei livelli dell'atomo di Rb.

Dettagli del setup:
* Eccitazione: laser [**Toptica DL PRO**](https://www.toptica.com/products/tunable-diode-lasers/ecdl-dfb-lasers/dl-pro) da 780nm regolabile in frequenza tramite un voltaggio di controllo;
* Parte della luce del laser viene inviata su un Fabry-Perot per la stima della lunghezza d'onda (segnale in V proporzionale alla potenza, scala non rilevante ai fini della misura);
* La fotoluminescenza viene misurata con un fotomoltiplicatore amplicato (il setup fornisce un voltaggio ma la scala assoluta della misura in termini di potenza è ignota). Notare: il segnale è negativo perché misurato dopo un preamplificatore invertente.

```python
import numpy as np
import matplotlib.pyplot as plt

# Lettura semplice skippando header
data = np.loadtxt('dati_Rb.txt', delimiter=',', skiprows=1)

plt.subplot(2,1,1)
plt.plot(data[:,0],data[:,1],'.')
plt.ylabel('Fotomol. [V]')

plt.subplot(2,1,2)
plt.plot(data[:,0],data[:,2],'-')
plt.xlabel('Voltaggio [V]')
plt.ylabel('Trasmissione[mW]')
```
